```python
import pandas as pd
import warnings as w
w.filterwarnings("ignore")
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv('titanic.csv')
df
```

```python
input=df.drop(columns=['PassengerId','Survived' ,'Name','SibSp','Parch','Ticket','Cabin' ,'Embarked'])
output=df["Survived"]
```

```python
decode=LabelEncoder()
input["Sex"] = decode.fit_transform(input["Sex"])
input
```

```python
input_a = input.drop(columns=["Sex"])
input_a.isnull().sum()
```

```python
input_a
```

```python
input["Age"].fillna(0, inplace =True)
```

```python
input_a.isnull().sum()
```

```python
input_a_train, input_a_test, output_train, output_test = train_test_split(input_a, output, train_size=0.8)
model = DecisionTreeClassifier()
model.fit(input_a_train, output_train)
predictions = model.predict(input_a_test)
```

```python
print("The Accuracy is  ",int ((model.score(input_a_train,output_train))*100),'%')
```
